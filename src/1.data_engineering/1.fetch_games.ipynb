# BUILD ALL GAMES FILE

Opens and combines all of the __year__ files we have in the data folder and writes the `all_games.csv` file.

```python
import glob

# gather all file paths
file_paths = glob.glob(f"{GAMES_DATA_PATH}/games_*.csv")
file_paths.sort()
file_paths

all_games_df_list = []
for file_path in file_paths:
    year_df = pd.read_csv(file_path)
    all_games_df_list.append(year_df)

all_games_df = pd.concat(all_games_df_list, ignore_index=True)
all_games_df.to_csv(f'{GAMES_DATA_PATH}/all_games.csv', index=False)
all_games_df
```
