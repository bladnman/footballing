```python
import pandas as pd
import numpy as np
```

```python
data1 = np.random.randint(10, size=(5, 3))
columns1 = ['Score A', 'Score B', 'Score C']
df = pd.DataFrame(data=data1, columns=columns1)

df
```

```python
data2 = np.random.randint(10, size=(5, 1))
df['Score D'] = data2

df
```

```python
data3 = np.random.randint(10, size=(5, 2))
columns3 = ['Score E', 'Score F']
df_add = pd.DataFrame(data=data3, columns=columns3)
df = pd.concat([df, df_add], axis=1)

df
```
