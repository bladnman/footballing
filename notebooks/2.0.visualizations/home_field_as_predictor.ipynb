# Home Field Advantage

```python
import utils.game_utils as gu
import utils.plot as guplot

import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


%load_ext autoreload
%autoreload 2

nfld = gu.NFL_Data()
nfl_df = nfld.data_by_game()
samp_df = nfl_df[['year', 'win']]
win_perc_year_df = samp_df.groupby('year').sum() / samp_df.groupby('year').count() * 100
win_perc_df = win_perc_year_df.reset_index()
guplot.bar_percent_plot_v2(win_perc_df, 
                           x='year',
                           y='win',
                           suptitle='Is being at home a predictor of victory? : 2010-2021', 
                           title='Home Field Advantage?')
```
